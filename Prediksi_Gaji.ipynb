# Prediksi Gaji 8 Tahun Kerja

### import library

Bagian ini merupakan impor library yang diperlukan untuk melakukan manipulasi data, visualisasi, dan regresi linear.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

### membaca dataset

Kode ini digunakan untuk membaca dataset dalam format CSV dan membagi data ke dalam atribut (x) dan label (y).

```python
dataset = pd.read_csv('Data Gaji.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

### memecah data

Kode ini memecah data menjadi data training dan data testing dengan rasio 80:20.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

### fitting regresi linear

Kode ini melakukan fitting model regresi linear pada data training.

```python
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

### memprediksi data baru

Kode ini menggunakan model regresi linear yang telah dilatih untuk memprediksi pendapatan berdasarkan masa kerja baru (8 tahun).

```python
new_data = np.array([[8]])  # Contoh data baru dengan masa kerja 8 tahun
predicted_salary = regressor.predict(new_data)
print("Prediksi Pendapatan untuk Masa Kerja 8 tahun: ", predicted_salary)
```

### visualisasi hasil regresi (data training dan data testing)

Kode-kode ini digunakan untuk membuat plot yang memvisualisasikan hubungan antara masa kerja dan pendapatan. Terdapat dua plot, yaitu plot untuk data training dan plot untuk data testing. Data asli ditampilkan sebagai titik-titik dalam scatter plot, sementara regresi linear ditampilkan sebagai garis dalam plot.

```python
plt.scatter(x_train, y_train, color='blue', label='Data Asli')
plt.plot(x_train, regressor.predict(x_train), color='red', label='Regresi Linear')
plt.title('Masa Kerja vs Pendapatan (Training)')
plt.xlabel('Masa Kerja')
plt.ylabel('Pendapatan')
plt.legend()
plt.show()
```

```python
plt.scatter(x_test, y_test, color='green', label='Data Asli')
plt.plot(x_test, regressor.predict(x_test), color='black', label='Regresi Linear')
plt.title('Masa Kerja vs Pendapatan (Testing)')
plt.xlabel('Masa Kerja')
plt.ylabel('Pendapatan')
plt.legend()
plt.show()
```
